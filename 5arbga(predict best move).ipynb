```python
#this function store the possible moves for each rock
def possible_moves(i,j):
    from collections import defaultdict
    d=defaultdict(list)
    if(i==0 and j==0):
        d[0].append(0)
        d[0].append(1)
        d[1].append(1)
        d[1].append(0)
        d[2].append(1)
        d[2].append(1)
    elif(i==0 and j==2):
        d[0].append(0)
        d[0].append(1)
        d[1].append(1)
        d[1].append(2)
        d[2].append(1)
        d[2].append(1)
    elif(i==2 and j==0):
        d[0].append(1)
        d[0].append(0)
        d[1].append(2)
        d[1].append(1)
        d[2].append(1)
        d[2].append(1)
    elif(i==2 and j==2):
        d[0].append(1)
        d[0].append(2)
        d[1].append(2)
        d[1].append(1)
        d[2].append(1)
        d[2].append(1)
    elif(i==0 and j==1):
        d[0].append(0)
        d[0].append(0)
        d[1].append(0)
        d[1].append(2)
        d[2].append(1)
        d[2].append(1)
    elif(i==2 and j==1):
        d[0].append(2)
        d[0].append(0)
        d[1].append(2)
        d[1].append(2)
        d[2].append(1)
        d[2].append(1)
    elif(i==1 and j==0):
        d[0].append(0)
        d[0].append(0)
        d[1].append(2)
        d[1].append(0)
        d[2].append(1)
        d[2].append(1)
    elif(i==1 and j==2):
        d[0].append(0)
        d[0].append(2)
        d[1].append(2)
        d[1].append(2)
        d[2].append(1)
        d[2].append(1)
    return d
```

```python
# if we have 3 rocks in the same line that mean the game is over 
# The value -10 means that the player beat the computer 
# The value 10 means that the player lost to the computer 
def evaluate(b,pc,op) :
   
    for row in range(3) :    
        if (b[row][0] == b[row][1] and b[row][1] == b[row][2]) :       
            if (b[row][0] == pc) :
                return 10
            elif (b[row][0] == op) :
                return -10
 
   
    for col in range(3) :
      
        if (b[0][col] == b[1][col] and b[1][col] == b[2][col]) :
         
            if (b[0][col] == pc) :
                return 10
            elif (b[0][col] == op) :
                return -10
 
    if (b[0][0] == b[1][1] and b[1][1] == b[2][2]) :
     
        if (b[0][0] == pc) :
            return 10
        elif (b[0][0] == op) :
            return -10
 
    if (b[0][2] == b[1][1] and b[1][1] == b[2][0]) :
     
        if (b[0][2] == pc) :
            return 10
        elif (b[0][2] == op) :
            return -10
 
   
    return 0
```

```python
def minimax(damma, depth, Max,alpha,beta,pc,op):

    score = evaluate(damma,pc,op)

    if (score == 10): 
        return score,depth

    if (score == -10):    
        return score,depth
    
    
    if (depth == 5):
        return 0,depth

    if (Max):
        dep=10
        best = -1000
        for i in range(2,-1,-1):
        
            for j in range(2,-1,-1):
            
                
                if (damma[i][j]==pc and ((i==1) and (j==1))==False):
                    v=possible_moves(i,j)
                    for x in range(3):

                        y=v[x][0]
                        z=v[x][1]
                        if (damma[y][z]=='_'):
                            damma[i][j] = '_'
                            damma[y][z] = pc
                            val_best,val_dep=minimax(damma, depth+1,not(Max),alpha,beta,pc,op)
                          
                            if(best==val_best):
                                    dep=min(dep,val_dep)
                            elif(best<val_best):
                                    dep=val_dep
                            best = max(best,val_best)
                            alpha = max(alpha, best)
                        

                            damma[i][j] = pc
                            damma[y][z] = '_'
                  

                
                elif(damma[i][j]==pc and ((i==1) and (j==1))==True):
                    for s in range(3):
        
                        for k in range(3):
            
                          if (damma[s][k]=='_' and ((s==1) and (k==1))==False):
                                damma[1][1] = '_'
                                damma[s][k] = pc
                                val_best,val_dep=minimax(damma, depth+1,not(Max),alpha,beta,pc,op)
                             
                                if(best==val_best):
                                    dep=min(dep,val_dep)
                                elif(best<val_best):
                                     dep=val_dep
                                best = max(best,val_best)
                                alpha = max(alpha,best)
                                damma[1][1] = pc
                                damma[s][k] = '_'
                             

                
            
        
        return best,dep
        
    

   
    else:
    
        best = 1000
        dep=10
      
        for i in range(2,-1,-1):
        
            for j in range(2,-1,-1):
            
               
                if (damma[i][j]==op and ((i==1) and (j==1))==False):
                    v=possible_moves(i,j)
                    for x in range(3):
                        y=v[x][0]
                        z=v[x][1]
                      
                        if (damma[y][z]=='_'):
                            damma[i][j] = '_'
                            damma[y][z] =op
                         
                            val_best,val_dep=minimax(damma, depth+1,not(Max),alpha,beta,pc,op)
                          
                            if(best==val_best):
                                    dep=min(dep,val_dep)
                            elif(best>val_best):
                                    dep=val_dep                           

                            best = min(best,val_best)
                            beta = min(beta,best)
                   
                            damma[i][j] = op
                            damma[y][z] = '_'
                     
                
                elif (damma[i][j]==op and ((i==1) and (j==1))==True):
                    for s in range(3):
        
                       for k in range(3):
            
                            if (damma[s][k]=='_' and ((s==1) and (k==1))==False):
                                damma[1][1] = '_'
                                damma[s][k] = op
                                val_best,val_dep=minimax(damma, depth+1,not(Max),alpha,beta,pc,op)
                              
                                if(best==val_best):
                                    dep=min(dep,val_dep)
                                elif(best>val_best):
                                    dep=val_dep
                                else:
                                    dep=dep
                            
                                best = min(best,val_best)
                                beta = min(beta,best)
                                damma[1][1] = op
                                damma[s][k] = '_'
                                                         

        return best,dep
    
```
